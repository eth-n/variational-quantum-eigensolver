```python
import numpy

from openfermion.hamiltonians import MolecularData
from openfermion.ops import FermionOperator
from openfermion.transforms import jordan_wigner

from openfermionpyscf import run_pyscf

from pyscf import gto, scf, df
```

```python
geometry = [['H', [0, 0, 0]], 
            ['H', [0, 0, 0.741]]]
basis = 'sto-3g'
charge = 0
multiplicity = 1
```

```python
molecule = MolecularData(geometry,
                         basis,
                         multiplicity,
                         charge)
```

```python
run_pyscf(molecule, run_scf=True)
print(molecule.get_integrals())
```

```python
hamiltonian_interaction_op = molecule.get_molecular_hamiltonian()
```

```python
for term in hamiltonian_interaction_op:
    ham_fermion_op = FermionOperator(term, hamiltonian_interaction_op[term])
#     print(ham_fermion_op)
    jw_ham_qubit_op = jordan_wigner(ham_fermion_op)
    print(jw_ham_qubit_op)
    print()
```
